```python
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_california_housing

data = fetch_california_housing()
data
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
df['Price'] = data.target
df.head()
```

```python
df.corr() #tocheck correlation exit between values
```

```python
!pip
```

```python
from statmodels.stats.outliers_influence import variance_inflation_factor
import numpy as np
```

```python
X = df.drop(columns=['Price'])
y = df['Price']
```

```python
X
```

```python
y
```

```python
X.columns
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
X_scaled =scalar.fit_transform(X)
X_scaled
```

Check for Multicollinearity




```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor
import numpy as np

vif_data = pd.DataFrame()
vif_data['Features'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X_scaled, i) for i in range(X_scaled.shape[1])]
vif_data
```

```python
X = X[['MedInc','HouseAge','AveRooms','AveBedrms','Population','AveOccup','Latitude','Longitude']]
X_scaled =scalar.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y, test_size=0.2,random_state=1)
```

```python
X_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression() #initialize
model.fit(X_train,y_train)#train

y_pred = model.predict(X_test)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
model.predict(X_test)#test
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
print("Mean absolute error: %.2f" % mean_absolute_error(y_test, y_pred))
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
print("Root mean squared error: %.2f" % np.sqrt(mean_squared_error(y_test, y_pred)))
print("R2-score: %.2f" % r2_score(y_test, y_pred))
```
